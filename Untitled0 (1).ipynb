```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Online Retail.csv', encoding='ISO-8859-1')

# Convert InvoiceDate to datetime with the correct format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d-%m-%Y %H:%M')

# Check for missing values
print(df.isnull().sum())

# Drop rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Create a TotalSales column
df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Extract year, month, and day from InvoiceDate
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day

# Display the first few rows
print(df.head())
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Aggregate sales by month
monthly_sales = df.groupby(['Year', 'Month'])['TotalSales'].sum().reset_index()

# Create a date column
monthly_sales['Date'] = pd.to_datetime(monthly_sales[['Year', 'Month']].assign(DAY=1))

# Set Date as the index
monthly_sales.set_index('Date', inplace=True)

# Plot the sales data
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['TotalSales'])
plt.title('Monthly Sales')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()

# Fit an ARIMA model
model = ARIMA(monthly_sales['TotalSales'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast the next 6 months
forecast = model_fit.forecast(steps=6)
print(forecast)
```

```python
# Aggregate product sales by month
product_sales = df.groupby(['StockCode', 'Year', 'Month'])['Quantity'].sum().reset_index()

# Example: Forecast demand for a specific product (e.g., StockCode = '85123A')
product = product_sales[product_sales['StockCode'] == '85123A']

# Create a date column
product['Date'] = pd.to_datetime(product[['Year', 'Month']].assign(DAY=1))

# Set Date as the index
product.set_index('Date', inplace=True)

# Plot the product demand
plt.figure(figsize=(10, 6))
plt.plot(product['Quantity'])
plt.title('Product Demand for StockCode 85123A')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.show()

# Fit an ARIMA model for the product
model = ARIMA(product['Quantity'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast the next 6 months
forecast = model_fit.forecast(steps=6)
print(forecast)
```
